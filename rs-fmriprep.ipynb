### fmri preprocessing using fmriprep

```python
import bids
import os
import matplotlib.pyplot as plt
from nilearn import image as nimg
from nilearn import plotting as nplot
```

```python
#for inline visualization in jupyter notebook
%matplotlib inline 
prefix = "/home/yat-lok/workspace/BIDS_data"
data_path = prefix + "/derivatives/fmriprep/"

#Base directory for fmriprep output
fmriprep_dir = data_path
layout= bids.BIDSLayout(fmriprep_dir, config=['bids','derivatives'])
layout.get_subjects()
layout.get_tasks()

# get the T1w files
T1w_files = layout.get(subject='122817', datatype='anat',
					   extension='.nii.gz',
					   return_type='file')
t1 = T1w_files[0]
t1_img = nimg.load_img(t1)
nplot.plot_anat(t1_img)
```
